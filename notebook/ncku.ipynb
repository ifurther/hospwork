```python
html = etree.HTML(g_total_old.text)
```

```python
html.cssselect('table#table4')
```

```python
divs = html.xpath('//p[@align="left"]') #html.xpath('//div[position()<3]')
for div in divs:
    ass = div.findall('a')  # 这里只能找到:div->a, 找不到:div->p->a
    for a in ass:
        if a is not None:
            #print(dir(a))
            print(a.text, url_base_old+"/"+a.attrib.get('href')) #文档(DOM)元素(Element)的属性：text, attrib
```
