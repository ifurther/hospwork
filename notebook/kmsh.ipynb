```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
from urllib.parse import urlparse
```

```python
url_base='https://www.kmsh.org.tw/affairs/6300_file/HRDataBase/'
url_work_table='index.asp'
url=url_base+url_work_table
g=requests.get(url)
soup=BeautifulSoup(g.content, 'html.parser')
```

```python
all_link_list = soup.find_all('a',{'target':"_self"})
```

```python
for link in all_link_list:
    link_text = link.get('href')
    if 'cntupd' in link_text:
        print(link_text.replace('cntupd.asp',url_base+'index_data.asp'))
        
```
